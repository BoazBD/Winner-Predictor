```python
import pandas as pd
pd.set_option('display.max_rows', 1000)
import re

def parse_hebrew(text):
     # Remove Unicode control characters
    text = re.sub(r'[\u202A\u202C\u202E]', '', text)
    # Switch opening and closing parentheses
    text = text.replace("(", "TEMP_PLACEHOLDER")
    text = text.replace(")", "(")
    text = text.replace("TEMP_PLACEHOLDER", ")")
    return text
```
